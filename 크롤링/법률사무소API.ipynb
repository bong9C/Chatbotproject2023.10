```python
lat_list, lng_list = [], []
for i in add.index:
    base_url = 'https://dapi.kakao.com/v2/local/search/address.json'
    header = {'Authorization': f'KakaoAK {kakao_key}'}
    url = f'{base_url}?query={quote(df["주소"][i])}'
    result = requests.get(url, headers=header).json()
    result['documents'][0].keys()
    lat_list.append(float(result['documents'][0]['y']))
    lng_list.append(float(result['documents'][0]['x']))
```

```python
lng_list
```

```python
add['위도'] = lat_list
add['경도'] = lng_list
```

```python
import folium
```

```python
df = pd.read_csv('data/공단.csv')
df
```

```python
map = folium.Map(location=[df.위도.mean(), df.경도.mean()], zoom_start=8)  # Center position
for i in df.index:
    folium.Marker(
        location=[df.위도[i], df.경도[i]],       
        tooltip=df.조직명[i],
        popup=folium.Popup(df.주소[i]+'<br>'+df.전화번호[i],max_width=200)

    ).add_to(map)

title_html = '<h3 align="center" style="font-size:20px">전국 대한법률구조 공단</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```

```python
filename = os.path.join('../static', 'img/공단map.html')
map.save(filename)
```
